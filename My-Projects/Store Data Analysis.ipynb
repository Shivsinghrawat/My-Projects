# Data Visualization

# 1-  Gender Distribution

```python
# Count the number of occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Create the pie chart
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')

# Set title
plt.title('Gender Distribution')

# Show the chart
plt.show()

```

Shopping behavior analysis revealed a gender distribution of 69.4% women and 30.6% men.

# 2 - Monthly sales Trend

```python
monthly_sales = df.groupby('Month')['Amount'].sum()
monthly_sales
```

```python
# Group the data by month and calculate the total sales amount for each month
monthly_sales = df.groupby('Month')['Amount'].sum()

# Create the line chart
plt.plot(monthly_sales.index, monthly_sales.values)

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.title('Monthly Sales Trend')

# Show the chart
plt.show()
```

Analyzed monthly sales trends from Janto Dec, revealing insights into patterns and fluctuations in sales performance throughout the year.

# 3 - Age Group Distribution

```python
# Group by 'Age Group' and calculate the count
grouped_data = df.groupby('Age group').size()

print(grouped_data)
```

```python
# Count the number of occurrences of each age group
age_group_counts = df['Age group'].value_counts()

# Create the bar chart
plt.bar(age_group_counts.index, age_group_counts.values)

# Set labels and title
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution')

# Show the chart
plt.show()
```

Explored age group distribution, categorizing individuals into adult, teenager, and senior segments, to gain insights into the demographic composition and target audience representation.

# 4 - Total Sales Amount by Category

```python
categories = np.array(df['Category'])
category_sales = {}
for category in np.unique(categories):
    sales = np.sum(amount_array[categories == category])
    category_sales[category] = sales
print("Total Sales Amount by Category:")
category_sales
```

```python
# Group the data by category and calculate the total sales amount for each category
category_sales = df.groupby('Category')['Amount'].sum()

# Create the bar chart
plt.bar(category_sales.index, category_sales.values)

# Set labels and title
plt.xlabel('Category')
plt.ylabel('Sales Amount')
plt.title('Category Sales')

# Rotate x-axis labels if needed
plt.xticks(rotation=50)

# Show the chart
plt.show()
```

Analyzed sales by category, providing insights into the performance and contribution of different product categories to overall sales revenue.

# 5 - Status Distribution

```python
# Count the number of occurrences of each status
status_counts = df['Status'].value_counts()

# Create the pie chart
plt.pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%')

# Set title
plt.title('Status Distribution')

# Show the chart
plt.show()
```

Analyzed the distribution of order statuses, including refund, cancel, return, and delivered, to gain insights into the frequency and proportion of each status within the dataset.

# 6 - Top 5 States with Highest Sales Amount

```python
category_sales = df.groupby('ship-state')['Amount'].sum()
top_state = category_sales.nlargest(5)
top_state
```

```python
sorted_sales = category_sales.sort_values(ascending=False)

# Extract the top 5 states with highest sales
top_states = sorted_sales.head(5)

# Create a horizontal bar plot
plt.barh(top_states.index, top_states.values)
plt.xlabel('Sales Amount')
plt.ylabel('State')
plt.title('Top 5 States with Highest Sales Amount')

# Display the plot
plt.show()
```

Identified the top 5 states with the highest sales amount, shedding light on the regions that contribute the most to overall sales revenue.
